Q1. **Greedy vs. Non-Greedy**:

ANS->

- **Greedy**: Matches as much as possible while still allowing the overall pattern to match.
- **Non-Greedy (Lazy)**: Matches as little as possible while still allowing the pattern to match.

To transform a greedy pattern into a non-greedy one, add a `?` after the quantifier (e.g., `*?`, `+?`, `??`).

Q2. **When Greedy vs. Non-Greedy Matters**:

ANS->

- Greedy vs. Non-Greedy matters when a pattern contains multiple possible matches and you want to control how much is consumed by a quantifier.
- If only a greedy match is available, you can still use it and adjust your pattern or logic accordingly.

Q3. **Non-Tagged Group in Simple Match**:

ANS->

- In simple matches without replacement, using a non-tagged group typically doesn't make a practical difference.

Q4. **Scenario with Non-Tagged Category**:

ANS->

- Using a non-tagged category can impact the outcome when capturing and referencing group content separately from the whole match, as it changes the group numbering.

Q5. **Look-Ahead Without Consuming**:

ANS->

- Look-ahead condition could be used to validate if a certain condition follows without actually consuming those characters.

Q6. **Positive vs. Negative Look-Ahead**:

ANS->

- **Positive Look-Ahead `(?!...)`**: Asserts that a sub-pattern doesn't follow the main pattern.
- **Negative Look-Ahead `(?=...)`**: Asserts that a sub-pattern does follow the main pattern.

Q7. **Named Groups vs. Numbered Groups**:

ANS->

- Referring to groups by name improves code readability and makes the code more self-documenting.

Q8. **Identifying Repeated Items Using Named Groups**:

ANS->

- Yes, you can use named groups and regular expressions to identify repeated items in a target string.

Q9. **Scanner Interface vs. re.findall**:

ANS->

- The Scanner interface in Python can provide more control during parsing, allowing you to perform custom actions while scanning the string, unlike `re.findall()` which focuses on pattern matching.

ANS->

- No, a Scanner object does not have to be named "scanner". You can choose any valid variable name you prefer.
