Q1. **Text vs. Binary Files**:

ANS->

Text files contain human-readable characters encoded in a specific character encoding (e.g., ASCII, UTF-8), and often have lines separated by newline characters. Binary files, on the other hand, contain a sequence of bytes without a specific character encoding, making them suitable for storing non-text data like images, audio, or program executables.

Q2. **Scenarios for Text and Binary Files**:

ANS->

- **Text Files**: Text files are suitable for storing and working with human-readable content, such as configuration files, logs, and structured data in CSV or JSON format.
- **Binary Files**: Binary files are used for non-text data like images, videos, audio, and proprietary data formats, where maintaining the exact sequence of bytes is critical.

Q3. **Issues with Binary Operations on Python Integers**:

ANS->

Issues can arise with endianness (byte order) and platform-specific integer sizes when using binary operations to read and write Python integers directly to disk. These issues can cause compatibility problems when reading the data on different platforms.

Q4. **Benefits of 'with' Keyword**:

ANS->

Using the `with` keyword (context manager) for file operations ensures that the file is properly closed after the block of code is executed, even if an exception occurs. It simplifies resource management and reduces the risk of resource leaks.

Q5. **Trailing Newline and Appending Newline**:

ANS->

- Python does not include the trailing newline character when reading a line of text using methods like `readline()`.
- By default, Python appends a newline character (`'\n'`) to the end of each line when you use the `write()` method to write a line of text to a file.

Q6. **Random-Access Operations**:

ANS->

The `seek()` and `tell()` methods enable random-access operations in files. `seek()` moves the file pointer to a specific position, allowing you to read or write at that position, and `tell()` returns the current position of the file pointer.

Q7. **Usage of struct Package**:

ANS->

The `struct` package is used for packing and unpacking binary data in fixed formats, especially when dealing with low-level binary data structures, such as network protocols and file formats.

Q8. **When to Use Pickling**:

ANS->

Pickling is best for serializing and deserializing complex Python objects, preserving their full structure, including custom classes, functions, and data.

Q9. **When to Use shelve Package**:

ANS->

The `shelve` package is useful when you want to store Python objects, including dictionaries and custom objects, persistently in a database-like format. It provides a convenient way to save and retrieve objects with keys.

Q10. **Special Restriction with shelve Package**:

ANS->

The `shelve` package stores keys as strings, which means that keys you use for storing and retrieving data must be strings. Other data dictionaries in Python (e.g., regular dictionaries) allow a wider range of keys, including various data types.
