Q1. Compare and contrast the float and Decimal classes&#39; benefits and drawbacks.

ANS->

 **Comparison of float and Decimal Classes**:
   - **Benefits of `Decimal` Class**: It provides arbitrary precision decimal arithmetic, suitable for financial calculations where exact decimal representation is crucial.
   - **Drawbacks of `Decimal` Class**: It's slower and requires more memory compared to `float`. Its operations may involve more computational overhead due to the precision handling.

different internal states?

ANS->

`Decimal('1.200')` and `Decimal('1.2')` are two different objects. They represent the same value, but the internal states may differ due to the way floating-point values are converted to `Decimal` objects.

ANS->

It's preferable to start a `Decimal` object with a string rather than a floating-point value to avoid potential precision loss caused by the conversion of floating-point values to decimals.
