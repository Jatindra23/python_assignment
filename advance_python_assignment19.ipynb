Q1. **Class and Instances Relationship**:

ANS->

The relationship between a class and its instances is a one-to-many partnership. A class defines a blueprint or template for creating objects, and instances are individual objects created from that class.

Q2. **Data Held Only in an Instance**:

ANS->

Instance-specific data is held in instance variables. These variables store data unique to each instance and are defined within the class's methods, often using the `self` keyword.

Q3. **Knowledge Stored in a Class**:

ANS->

A class stores knowledge about its attributes (variables) and behaviors (methods) that its instances share. It encapsulates a common structure and functionality that instances can utilize.

Q4. **Method vs. Regular Function**:

ANS->

A method is a function defined within a class that operates on its instances. It takes `self` as its first parameter to refer to the instance it's operating on. A regular function is not associated with any class and does not require an instance to operate.

ANS->

```python
class ChildClass(ParentClass):
    # Class definition
```

Q6. **Encapsulation in Python**:

ANS->

Python supports encapsulation to a certain extent by using naming conventions. Prefixing variables with a single underscore `_` implies they are meant to be treated as "protected" (not to be accessed outside the class) and using double underscore `__` makes them "private" (name-mangled to make them less accessible).

Q7. **Class Variable vs. Instance Variable**:

ANS->

A class variable is shared among all instances of a class, while an instance variable is specific to an individual instance. Class variables are defined outside of methods, while instance variables are defined within methods using `self`.

Q8. **Self in Method Definitions**:

ANS->

`self` is included as the first parameter in a class's method definitions to refer to the instance on which the method is being called. It's necessary to include `self` in order to access instance-specific data and methods within the method.

Q9. **Difference between \_\_add\_\_ and \_\_radd\_\_**:

ANS->

- `__add__`: Defines the behavior of the `+` operator when the object is on the left side.
- `__radd__`: Defines the behavior of the `+` operator when the object is on the right side.

Q10. **Use of Reflection Methods**:

ANS->

Reflection methods like `__getattr__` and `__setattr__` allow customization of attribute access and assignment. You may not need them if you're satisfied with the default behavior of attribute access and assignment.

Q11. **\_\_iadd\_\_ Method**:

ANS->

`__iadd__` is called for `+=` operations. It's used to modify the object in-place and is part of the augmented assignment protocol.

Q12. **\_\_init\_\_ Method Inheritance**:

ANS->

Yes, the `__init__` method is inherited by subclasses. If you need to customize its behavior within a subclass, you can override it by defining a new `__init__` method in the subclass. To call the parent class's `__init__` method, use `super().__init__(...)`.
